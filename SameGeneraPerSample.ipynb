# Find same genera in same sample

```python
# read matrix
#matrix1 = pd.read_csv('results/matrix_genus_only.csv', index_col=0)
```

```python
# prepare data
matrix1 = matrix.reset_index()
# split column values with pattern '--'
patt = matrix1['cluster'].str.split('--', expand=True)
matrix1['genus'] = patt[0]

# Aggregate by Genus and sum the scores
matrix_grouped = matrix1.groupby('genus').sum()
matrix_grouped.to_csv('results/matrix_genus_only_grouped.csv')
matrix_grouped.head()
```

```python
# Skip

matrix_grouped.iloc[:, 0:] = matrix_grouped.iloc[:, 0:].replace(1, 0)
matrix_grouped = matrix_grouped[matrix_grouped.iloc[:, 0:].sum(axis=1) > 0]

matrix_grouped
```

```python
# Calculate the sum of values for each column
column_sums = matrix_grouped.sum(axis=0)

# Filter the DataFrame to keep only columns with sum different from 0
matrix_filtered = matrix_grouped[column_sums[column_sums != 0].index]
matrix_filtered
```

```python
matrix_filtered.to_csv('results/matrix_genus_only_grouped_filtered.csv')
```

```python
# Sort columns by their sum
column_sums = matrix_filtered.sum(axis=0)
matrix_filtered_sorted_columns = matrix_filtered[column_sums.sort_values(ascending=False).index]

# Sort rows by their sum
row_sums = matrix_filtered.sum(axis=1)
matrix_filtered_sorted = matrix_filtered_sorted_columns.loc[row_sums.sort_values(ascending=False).index]

###
matrix_filtered_sorted.to_csv('results/matrix_genus_only_grouped_filtered_sorted.csv')
```
